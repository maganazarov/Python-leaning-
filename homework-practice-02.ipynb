```python
from time import time

def compare_functions(f1, f2, args):
    start = time()
    f1(*args)
    end = time()
    f1_time = end - start
    start = time()
    f2(*args)
    end = time()
    f2_time = end - start
    return f1_time, f2_time

def generate_args(func, n):
    if func.__name__ == 'calculate_diagonal':
        return (np.random.randint(low=0, high=255, size=(n, n)),)
    elif func.__name__ == 'multiset_equal':
        return (np.random.randint(low=0, high=255, size=n),
                np.random.randint(low=0, high=255, size=n))
    elif func.__name__ == 'find_max_element_after_zero':
        return (np.random.randint(low=0, high=255, size=n),)
    elif func.__name__ == 'convert_image':
        return (np.random.randint(low=0, high=255, size=(n, n, 3)), np.random.rand(3))
    elif func.__name__ == 'run_length_encoding':
        return (np.random.randint(low=0, high=255, size=n),)
    elif func.__name__ == 'pairwise_distance':
        return (np.random.randint(low=0, high=255, size=(n, 100)),
                np.random.randint(low=0, high=255, size=(n, 100)))
```

```python
f, axes = plt.subplots((n_plots + 1) // 2, 2, figsize=(20, 20))
for i in range(n_plots):
    time_not_vectorized = []
    time_vectorized = []
    time_scipy = []
    f_vec, f_not_vec = functions_dict['vectorized'][i], functions_dict['not_vectorized'][i]
    for n in data_size:
        args = generate_args(f_vec, n)
        times = compare_functions(f_vec, f_not_vec, args)
        time_vectorized.append(times[0])
        time_not_vectorized.append(times[1])
        if f_vec.__name__ == 'pairwise_distance':
            start = time()
            cdist(*args)
            end = time()
            time_scipy.append(end - start)
    ax = axes[i // 2, i % 2]
    ax.plot(data_size, time_not_vectorized, label=u"Невекторизованная " + f_not_vec.__name__)
    ax.plot(data_size, time_vectorized, label=u"Векторизованная " + f_vec.__name__)
    if f_vec.__name__ == 'pairwise_distance':
        ax.plot(data_size, time_vectorized, label=u"Векторизованная " + cdist.__name__)
    ax.set_xlabel(u"Длина массива")
    ax.set_ylabel(u"Время (с)")
    ax.grid()
    ax.legend()
f.show()
```

```python
def pairplot(df, target):
    ncol, nrow = 7, df.shape[1] // 7 + (df.shape[1] % 7 > 0)
    plt.figure(figsize=(ncol * 4, nrow * 4))

    for i, feature in enumerate(df.columns):
        plt.subplot(nrow, ncol, i + 1)
        plt.scatter(df[feature], target, s=10, marker='o', alpha=.6)
        plt.xlabel(feature)
        if i % ncol == 0:
            plt.ylabel('target')
```

```python
#  Your code here

import matplotlib.pyplot as plt
import numpy as np

from sklearn.linear_model  import (RidgeCV, LassoCV)

alphas = np.logspace(-6, 6, 50)
ridge_cv = RidgeCV(alphas=alphas)
ridge_cv.fit(x_train, y_train)
lasso_cv = LassoCV(alphas=alphas)
lasso_cv.fit(x_train, y_train)
print('Линейная регрессия c L-1 регуляризацией:',lasso_cv.score(x_test, y_test), sep='\n\t')
print('Линейная регрессия с L-2 регуляризацией:',ridge_cv.score(x_test, y_test), sep='\n\t')
```

Постройте график зависимости $R^2$ для обучающей и тестовой (на кросс-валидации) выборок в зависимости от значения гиперпараметра. Для Lasso также постройте график зависимости количества ненулевых весов.

```python
#  Your code here
from sklearn.model_selection import cross_validate
estimators = [Lasso, Ridge]
f, axes = plt.subplots(1, 2, figsize=(13, 5))
for i, estimator in enumerate(estimators):
    test_cv_scores = []
    train_cv_scores = []
    test_scores = []
    for a in alphas:
        regr = estimator(alpha=a)
        cv = cross_validate(regr, x_train, y_train, cv=10, scoring='r2', return_train_score=True)
        test_cv_scores.append(cv['test_score'].mean())
        train_cv_scores.append(cv['train_score'].mean())
        test_scores.append(regr.fit(x_train, y_train).score(x_test, y_test))
    ax = axes[i] 
    ax.set_title(estimator.__name__)
    ax.plot(np.log(alphas), test_cv_scores, label=u'Score on CV-test')
    ax.plot(np.log(alphas), train_cv_scores, label='Score on CV-train')
    ax.plot(np.log(alphas), test_scores, label='Score on test')
    ax.set_xlabel("log scaled alpha")
    ax.set_ylabel("R2 score")
    ax.grid()
    ax.legend()

f.show()
```
